```python
box = 0

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_eye.xml')

cap = cv2.VideoCapture(0)
_, img = cap.read()

def show_frame():
    global img, box
    while True:
        _, img = cap.read()
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(img, 1.3, 5)
        for (x,y,w,h) in faces:
            roi_gray = img[y:y+h, x:x+w]
            eyes = eye_cascade.detectMultiScale(roi_gray, 1.3, 5) # detect eyes
            width = np.size(roi_gray, 1) # get face frame width
            height = np.size(roi_gray, 0) # get face frame height
            for (x, y, w, h) in eyes:
                if y > height / 2:
                    pass
                eyecenter = x + w / 2  # get the eye center
                if eyecenter < width * 0.5:
                    left_eye = roi_gray[y:y + h, x:x + w]
                    left_eye = cv2.resize(left_eye, dsize=(100, 100))
                    cv2.imshow("eye", left_eye)
                    left_eye = np.array(left_eye).reshape(-1, 10000)
                    left_eye = left_eye.astype('float32')
                    left_eye /= 255
                    prediction = model.predict([left_eye])
                    box = np.argmax(prediction, axis = -1)
        interrupt = cv2.waitKey(10)
        if interrupt & 0xFF == 27: # esc key
            break

def tkinter_frame():
    global myCanvas, gaze
    frame = Tk()  
    frame.attributes('-fullscreen',True) 
    
    myCanvas = Canvas(frame)
    myCanvas.pack(fill=BOTH, expand=True)
    for key in map_box_to_coordinates_200.keys():
        x = map_box_to_coordinates_200[key]['x'][0]
        y = map_box_to_coordinates_200[key]['y'][0]
        box = myCanvas.create_rectangle(x, y, x+200, y+200, outline="#000000")
    gaze = myCanvas.create_rectangle(0, 0, 200, 200, outline="#000000", fill="#ff0000")
    frame.mainloop()

def pointer():
    global box
    x = map_box_to_coordinates_200[box]['x'][0]
    y = map_box_to_coordinates_200[box]['y'][0]
    while True:
        create_rectangle(x, y)
        
def create_rectangle(x, y): #center coordinates, radius
    global myCanvas, gaze
    try:
        myCanvas.delete(gaze)
    except:
        pass
    try:
        gaze = myCanvas.create_rectangle(x, y, x+200, y+200, outline="#000000", fill="#ff0000")
    except:
        pass
        

t1 = threading.Thread(target=show_frame)
t2 = threading.Thread(target=tkinter_frame)
t3 = threading.Thread(target=pointer)

# starting thread 1
t1.start()
# starting thread 2
t2.start()
t3.start()

# wait until thread 1 is completely executed
t1.join()
# wait until thread 2 is completely executed
t2.join() 
```
